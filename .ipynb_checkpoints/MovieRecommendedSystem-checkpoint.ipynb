```python
import pandas as pd
# from sklearn.metrics.pairwise import cosine_similarity
```

```python
movies = pd.read_csv('ml-latest-small/movies.csv')
movies.head()
```

```python
# tags = pd.read_csv('ml-latest-small/tags.csv')
# tags.head()
```

```python
ratings = pd.read_csv('ml-latest-small/ratings.csv')
ratings.head()
```

```python
df = pd.merge(movies, ratings)
df.head()
```

```python
df = df.drop(columns= ['timestamp'])
```

```python
df.head()
```

```python
userRatings = df.pivot_table(index=['userId'],columns=['title'],values='rating')
userRatings.head()
```

```python
corrMatrix = userRatings.corr()
corrMatrix.head()
```

```python
corrMatrix = userRatings.corr(method='pearson', min_periods=100)
corrMatrix.head()
```

```python
myRatings = userRatings.loc[1].dropna()
myRatings
```

```python
simMovies = pd.Series()

for i in range(0, len(myRatings.index)):
#     print("Adding similiar movies for " + myRatings.index[i] +"....")
    sims = corrMatrix[myRatings.index[i]].dropna()
    
    sims = sims.map(lambda x: x*myRatings[i]) # this gives rating to the new movies according to my ratings 
    
    simMovies = pd.concat([simMovies, sims]) # this produces a series of movies which can be recommended
```

```python
simMovies = simMovies.groupby(simMovies.index).sum()
simMovies.sort_values(inplace = True , ascending = False)

print(simMovies.head(10))

```

```python
recommendedMovies = simMovies[~simMovies.index.isin(myRatings.index)]
recommendedMovies.sort_values(inplace=True, ascending=False)
recommendedMovies.head(10)
# print(recommendedMovies)
```
